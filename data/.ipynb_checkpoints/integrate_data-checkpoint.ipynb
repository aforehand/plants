### incorporating pfaf data into usda data

```python
import pandas as pd
import numpy as np
```

```python
pfaf = pd.read_csv('pfaf_data.csv')
usda = pd.read_csv('all_native_plants.csv')
```

```python
usda = usda.rename({c: c.lower() for c in usda.columns}, axis=1)
```

```python
pfaf.shape
```

```python
usda.shape
```

```python
usda = usda.drop_duplicates(subset=['genus', 'species'])
usda.shape
```

```python
usda.reset_index(inplace=True)
```

```python
usda.to_csv('usda_data.csv', index=False)
```

```python
both = usda.merge(pfaf, on=['genus', 'species'])
```

```python
dupes = [c for c in both.columns if '_x' in c]
```

```python
dupes = [d[:-2] for d in dupes]
```

```python
for d in dupes:
    both[d] = both[[d+'_x', d+'_y']].any(axis=1)
```

```python
dupes = [c for c in both.columns if '_x' in c or '_y' in c]
both.drop(dupes, axis=1, inplace=True)
```

```python
for c in both.columns:
    print(c)
```

```python
[c for c in both.columns if 'pollinate' in c]
```

```python
both['pollinated by moths'] = both[['pollinated by moths and butterflies', 
                                    'pollinated by moths']].any(axis=1)
both['pollinated by butterflies'] = both[['pollinated by moths and butterflies', 
                                          'pollinated by butterflies']].any(axis=1)
```

```python
both['pollinated by insects'] = both[['pollinated by insects', 
                                      'pollinated by various insects']].any(axis=1)
```

```python
both.drop(['pollinated by moths and butterflies', 
           'pollinated by various insects'], axis=1, inplace=True)
```

```python
both['pollinated by flies'] = both[['pollinated by flies', 
                                    'pollinated by midges']].any(axis=1)
```

```python
both.drop('pollinated by midges', axis=1, inplace=True)
```

```python
both['pollinated by bees'] = both[['pollinated by bees', 
                                   'pollinated by bumblebees']].any(axis=1)
```

```python
both.drop('pollinated by bumblebees', axis=1, inplace=True)
```

```python
[c for c in both.columns if '(' in c]
```

```python
metals = ['b (boron)', 'co (cobalt)', 'cu (copper)', 'fe (iron)', 'mn (manganese)', 
          'na (sodium)', 'si (silicon)', 'zn (zinc)']
nutrients = ['ca (calcium)', 'k (potassium)', 'mg (magnesium)', 'p (phosphorus)', 
             's (sulfur)']
```

```python
both.rename({i: 'dynamic accumulator '+i for i in nutrients}, axis=1, inplace=True)
both.rename({i: 'hyperaccumulator '+i for i in metals}, axis=1, inplace=True)
```

```python
both[nutrients+['dynamic accumulator']].any(axis=1).value_counts()
```

```python
[c for c in both.columns if ('grow' in c) | ('lea' in c) | ('flower' in c)]
```

```python
both['fast growth'] = (both['growth rate']=='Rapid')|both['fast growth']
both['medium growth'] = (both['growth rate']=='Moderate')|both['medium growth']
both['slow growth'] = (both['growth rate']=='Slow')|both['fast growth']
```

```python
both.drop('growth rate', axis=1, inplace=True)
```

```python
both['flower time early spring'] = both[['flower mar', 'flower apr', 
                                         'flower time early spring']].any(axis=1)
both['flower time mid spring'] = both[['flower may', 'flower apr', 
                                         'flower time mid spring']].any(axis=1)
both['flower time late spring'] = both[['flower may', 'flower jun', 
                                         'flower time late spring']].any(axis=1)
both['flower time early summer'] = both[['flower jun', 'flower jul', 
                                         'flower time early summer']].any(axis=1)
both['flower time mid summer'] = both[['flower aug', 'flower jul', 
                                         'flower time mid summer']].any(axis=1)
both['flower time late summer'] = both[['flower aug', 'flower sep', 
                                         'flower time late summer']].any(axis=1)
both['flower time fall'] = both[['flower sep', 'flower oct', 'flower nov', 
                                         'flower time fall']].any(axis=1)
both['flower time winter'] = both[['flower dec', 'flower jan', 'flower feb', 
                                         'flower time fall']].any(axis=1)
```

```python
[c for c in both.columns if 'edible' in c]
```

```python
both['edible inner bark'] = both[['edible inner bark', 'bark']].any(axis=1)
both['edible roots'] = both[['edible roots', 'roots']].any(axis=1)
both['edible seeds'] = both[['edible seeds', 'seeds or nuts']].any(axis=1)
both['edible sap'] = both[['edible sap', 'sap']].any(axis=1)
both['edible leaves'] = both[['edible leaves', 'leaves']].any(axis=1)
both['edible flowers'] = both[['edible flowers', 'flowers']].any(axis=1)
both['edible fruit'] = both[['edible fruit', 'fruit']].any(axis=1)
both['edible stems'] = both[['edible stems', 'stem']].any(axis=1)
both['edible use oil'] = both[['edible oil', 'edible use oil']].any(axis=1)
```

```python
both.drop(['bark', 'roots', 'seeds or nuts', 'sap', 'leaves', 'flowers', 'fruit', 
           'stem', 'edible oil'], axis=1, inplace=True)
```

```python
both.to_csv('all_native_plants.csv', index=False)
```
